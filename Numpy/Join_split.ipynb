```python
a1=np.array([2,4,6,8])
a2=np.array([10,12,14,16])
print(np.concatenate([a1,a2]))
```

```python
arr1=np.array([[4,8],[12,16],[20,24]])
arr2=np.array([[5,10],[15,20],[25,30]])
print(np.concatenate([arr1,arr2],axis=0))
print(np.hstack([arr1,arr2]))
print(np.vstack([arr1,arr2]))
      
```
