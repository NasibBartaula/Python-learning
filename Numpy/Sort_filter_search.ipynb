```python
import numpy as np
a=np.array([5,4,3,2,6,7,9,8])
a=np.sort(a)
print(a)
```

```python
b=np.where(a==4)
print(b)
c=np.where(a%2==0)
print(c)
```

```python
ss=np.array([1,2,3,4,5,6,7])
z=np.searchsorted(ss,3)
print(z)
```
