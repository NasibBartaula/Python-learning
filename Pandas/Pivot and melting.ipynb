```python
import pandas as pd
dict={"id":[1,2,3],
     "name":["nasib","nitin","nirjal"],
      "houses":["red","green","blue"]}
a=pd.DataFrame(dict)
print(a)
print(a.pivot(index="id", columns="name", values="houses"))
```

```python
b=pd.DataFrame(dict)
print(b)
print(a.melt(id_vars="id",value_vars=["name","houses"]))
```
