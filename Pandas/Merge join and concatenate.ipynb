```python
import pandas as pd
a={"count":[1,2,3,4,5],
  "Name":["a","b","c","d","e"],
   "Salary":[10,20,10,20,30]
  }
b={"count":[1,2,3,4,5],
  "Name":["f","g","h","i","j"],
   "Salary":[20,87,45,76,45]
  }
n1=pd.DataFrame(a)
n2=pd.DataFrame(b)
```

```python
print(pd.merge(n1,n2,on = "count",how="left"))
```

```python
aa=pd.DataFrame(a)
bb=pd.DataFrame(b)
print(pd.concat([aa,bb]))
```
