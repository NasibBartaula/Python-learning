```python
import matplotlib.pyplot as plt
import pandas as pd
brands=["Nokia","Samsung","Apple","Redmi","Oneplus"]
c=["grey","blue","red","green","yellow"]
x=[20,50,10,70,30]
ex=[0,0,0.1,0,0]
plt.pie(x,labels=brands,colors=c,explode=ex,shadow=True,autopct="%i",startangle=360);
plt.savefig("PIE.png",facecolor="green",pad_inches=0.9)
plt.show()
```
