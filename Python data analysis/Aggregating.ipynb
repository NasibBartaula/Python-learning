```python
# NumPy Statistical & Aggregate Functions
# ======================================
#Create a NumPy array with elements [10,12,14,16,18,20].
#Find and print the maximum value of the array.
#Find and print the minimum value of the array.
#Find and print the sum of all elements.
#Find and print the total number of elements in the array.
#Find and print the mean (average) of the array.
#Find and print the cumulative sum of the array.
#Find and print the cumulative product of the array.

import numpy as np
a=np.array([10,12,14,16,18,20])
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.size(a))
print(np.mean(a))
print(np.cumsum(a))
print(np.cumprod(a))
```

```python
a=[200,140,350,220]
b=[22,31,24,16]
price=np.array(a)
quantity=np.array(b)
c=np.cumprod([price,quantity],axis=0)
print(c[1].sum())
```
